```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
class_names=['Banana','Apple','Avocado','Mango','Pawpaw','Orange','Lemon','Passionfruit']
```

```python
width=96
height=96
```

```python
from glob import glob
from keras import preprocessing 

#load_imagetype1
images_type1=[]
for image_path in glob('Banana/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type1.append(x)
    
images_type2=[]
for image_path in glob('Apple/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type2.append(x)
    
images_type3=[]
for image_path in glob('Avocado/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type3.append(x)
    
images_type4=[]
for image_path in glob('Mango/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type4.append(x)
    
images_type5=[]
for image_path in glob('Pawpaw/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type5.append(x)
    
images_type6=[]
for image_path in glob('Orange/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type6.append(x)
    
images_type7=[]
for image_path in glob('Lemon/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type7.append(x)
    
images_type8=[]
for image_path in glob('Passionfruit/*.*'):
    image=preprocessing.image.load_img(image_path,target_size=(width,height))
    x=preprocessing.image.img_to_array(image)
    images_type8.append(x)
    
```

```python
plt.figure(figsize=(12,8))
for i,x in enumerate(images_type1[:2]):
    plt.subplot(1,5,i+1)
    image=preprocessing.image.array_to_img(x)
    plt.imshow(image)
    
    plt.axis('off')
    plt.title('{} image'.format(class_names[0]))
    #show the plot
    plt.show()
```

```python
plt.figure(figsize=(12,8))
for i,x in enumerate(images_type5[:3]):
    plt.subplot(1,5,i+1)
    image=preprocessing.image.array_to_img(x)
    plt.imshow(image)
    
    plt.axis('off')
    plt.title('{} image'.format(class_names[4]))
    #show the plot
    plt.show()
```

```python
X_type_1=np.array(images_type1)
X_type_2=np.array(images_type2)
X_type_3=np.array(images_type3)
X_type_4=np.array(images_type4)

X_type_5=np.array(images_type5)
X_type_6=np.array(images_type6)
X_type_7=np.array(images_type7)
X_type_8=np.array(images_type8)

print(X_type_1.shape)
print(X_type_2.shape)
print(X_type_3.shape)
print(X_type_4.shape)
print(X_type_5.shape)
print(X_type_6.shape)
print(X_type_7.shape)
print(X_type_8.shape)
```

```python
X=np.concatenate((X_type_1,X_type_2),axis=0)

if len(X_type_3):
    X=np.concatenate((X,X_type_3),axis=0)

if len(X_type_4):
    X=np.concatenate((X,X_type_4),axis=0)
    
if len(X_type_5):
    X=np.concatenate((X,X_type_5),axis=0)
    
if len(X_type_6):
    X=np.concatenate((X,X_type_6),axis=0)
    
if len(X_type_7):
    X=np.concatenate((X,X_type_7),axis=0)
    
if len(X_type_8):
    X=np.concatenate((X,X_type_8),axis=0)

X=X/255 

    
print(X.shape)
print(X)
```

```python
from keras.utils import to_categorical

Y_type_1=[0 for item in enumerate(X_type_1)]
Y_type_2=[1 for item in enumerate(X_type_2)]
Y_type_3=[2 for item in enumerate(X_type_3)]
Y_type_4=[3 for item in enumerate(X_type_4)]

Y_type_5=[4 for item in enumerate(X_type_5)]
Y_type_6=[5 for item in enumerate(X_type_6)]
Y_type_7=[6 for item in enumerate(X_type_7)]
Y_type_8=[7 for item in enumerate(X_type_8)]

Y=np.concatenate((Y_type_1,Y_type_2),axis=0)

if len(Y_type_3):
    Y=np.concatenate((Y,Y_type_3),axis=0)

if len(Y_type_4):
    Y=np.concatenate((Y,Y_type_4),axis=0)
    
if len(Y_type_5):
    Y=np.concatenate((Y,Y_type_5),axis=0)
    
if len(Y_type_6):
    Y=np.concatenate((Y,Y_type_6),axis=0)
    
if len(Y_type_7):
    Y=np.concatenate((Y,Y_type_7),axis=0)
    
if len(Y_type_8):
    Y=np.concatenate((Y,Y_type_8),axis=0)

Y=to_categorical(Y,num_classes=len(class_names))
    
print(Y.shape)
print(Y)
    
```

```python
from keras.models import Sequential
from keras.layers.core import Activation , Dropout , Flatten , Dense
from keras.layers.convolutional import Convolution2D ,MaxPooling2D
from keras.optimizers import Adam

conv_1=16
conv_1_drop=0.2
conv_2=32
conv_2_drop=0.2
dense_1_n=1024
dense_1_drop=0.2
dense_2_n=512
dense_2_drop=0.2

epochs=30
batch_size=32
color_channels=3

def build_model(conv_1_drop=conv_1_drop,conv_2_drop=conv_2_drop,dense_1_n=dense_1_n,dense_1_drop=dense_1_drop,
               dense_2_n=dense_2_n,dense_2_drop=dense_2_drop):
    model=Sequential()
    
    model.add(Convolution2D(conv_1,(3,3),input_shape=(width,height,color_channels),activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(conv_1_drop))
    
    model.add(Convolution2D(conv_2,(3,3),activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(conv_2_drop))
    
    model.add(Flatten())
    
    model.add(Dense(dense_1_n,activation='relu'))
    model.add(Dropout(dense_1_drop))
 
    model.add(Dense(dense_2_n,activation='relu'))
    model.add(Dropout(dense_2_drop))
    
    model.add(Dense(len(class_names),activation='softmax'))
    
    model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
    
    return model

```

```python
np.random.seed(1) #for reproducability

#model with parameters
model=build_model()
model.summary()
```

```python
X
```

```python
model.fit(X,Y,epochs=epochs)
```

```python
#Predict using the model
type_1=preprocessing.image.load_img('Passionfruit/IMG_20190407_135142 - Copy.jpg',target_size=(width,height))
plt.imshow(type_1)
plt.show()

type_1_x=np.expand_dims(type_1,axis=0)
predictions=model.predict(type_1_x)
class_names[np.argmax(predictions)]
```
